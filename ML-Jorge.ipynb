```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['figure.figsize'] = (15.0, 12)

xgb.plot_importance(xgb_cls, importance_type="gain", values_format='{v:.2f}')
```

```python
# Se sacan los valores del plot importance y hacemos la media
# Como da un diccionario, se saca un array
importance = xgb_cls.get_booster().get_score(importance_type='gain')

importances_array = np.array(list(importance.values()))


importance_mean = np.mean(importances_array)
importance_mean
```

```python
# Se sacan los nombres de la variables con .keys()

importance_variables = np.array(list(importance.keys()))

# Se eliminan las variables que estÃ©n por debajo de la media
importance_variables = importance_variables[importances_array >= importance_mean]

# Se aÃ±ade lo que en un futuro serÃ¡ nuestro target
importance_variables = np.append(importance_variables, 'Diagnosis')

variables = importance_variables
df_importance = df[variables]
df_importance
```

## Hora de buscar los hyperparametros que funcionen mejor con RandomizedSearchCV

Los hyperparametros con los que se va a trabajar son:

- `n_estimators`: El nÃºmero de arboles de nuestro modelo.

- `max_depth`: Con esto observaremos cuanto se debe profundicar en cada Ã¡rbol.

- `subsample`: Escogemos una proporciÃ³n o fracciÃ³n de los datos usados por cada Ã¡rbol.

- `learning_rate`: Le restamos el peso que puedan tener los Ã¡rboles para evitar el overfitting.

- `alpha`: Le restamos el peso que puedan tener las hojas para evitar el overfitting.



1. Modelo con todas las variables

```python
# Declaramos X e y correspondientes al primer data frame con todas las variables
X_org = df_org.drop(columns=['Diagnosis'])
y_org = df_org['Diagnosis']
```

```python
# Realizamos el split, y declaramos el XGBClassifier

X_train_org, X_test_org, y_train_org, y_test_org = train_test_split(X_org, y_org, test_size=.2, random_state=42)
xgb_cls_org = xgb.XGBClassifier(random_state = 42)
```

```python
# Comprobamos el split

print(y_test_org.mean())
print(y_train_org.mean())
print(df_org['Diagnosis'].mean())
```

```python
# Creamos los parametros del RandomizeSearchCV

from sklearn.model_selection import RandomizedSearchCV

rs_param_grid = {
    'n_estimators': [10, 25, 40],
    'max_depth': list((range(3,12))),
    'subsample': [0.25,0.5,0.75,1],
    'learning_rate': np.linspace(0.01,0.5, 1),
    'alpha': [0,0.001, 0.01,0.1,1]
    }

```

```python
# Hacemos el RandomizedSearchCV

xgb_rs_org = RandomizedSearchCV(estimator=xgb_cls_org,param_distributions=rs_param_grid, 
                                cv=3, n_iter=500, random_state=42)
```

```python
# Entrenamos nuestro modelo

xgb_rs_org.fit(X_train_org, y_train_org)
```

```python
xgb_rs_org.best_params_
```

```python
xgb_cls_org.set_params(n_stimators=10, max_depth= 4, subsample=0.5, learning_rate=0.01, alpha= 0.01)
xgb_cls_org.fit(X_train_org, y_train_org)
```

```python
print(confusion_matrix(y_train_org, xgb_cls_org.predict(X_train_org)))
print(accuracy_score(y_train_org, xgb_cls_org.predict(X_train_org)))
print(recall_score(y_train_org, xgb_cls_org.predict(X_train_org)))
print(precision_score(y_train_org, xgb_cls_org.predict(X_train_org)))
```

```python
print(confusion_matrix(y_test_org, xgb_cls_org.predict(X_test_org)))
print(accuracy_score(y_test_org, xgb_cls_org.predict(X_test_org)))
print(recall_score(y_test_org, xgb_cls_org.predict(X_test_org)))
print(precision_score(y_test_org, xgb_cls_org.predict(X_test_org)))
```
